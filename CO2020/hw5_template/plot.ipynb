```python
import numpy as np
import matplotlib.pyplot as plt
import subprocess
```

```python
#dt = 1.25e-7
nx = np.array([10, 15, 20, 30, 40, 50, 60, 70, 80, 90, 100])
err_1 = np.array([0.027789, 0.004677, 0.006970, 0.007206, 0.008082, 0.008699, 0.009137, 0.009462, 0.009710, 0.009907, 0.010066])
#dt = 3.125e-6
err_2 = np.array([0.027632, 0.004640, 0.006885, 0.007163, 0.008049, 0.008670, 0.009111, 0.009436, 0.009686, 0.009883, 0.010042])
nx = np.log10(nx)
err_1 = np.log10(err_1)
err_2 = np.log10(err_2)
plt.plot(nx, err_1, 'r', label='dt = 1.25e-7')
plt.plot(nx, err_2, 'b.', label='dt = 3.125e-6')
# increase font size of axis labels and legend
plt.rcParams.update({'font.size': 12})
plt.xlabel('$log(nx)$')
plt.ylabel('$log(error)$')
plt.legend()
plt.show()
```

```python
print((err_1[1]-err_1[0])/(nx[1]-nx[0]))
```

```python
nx = []
err_1 = []
# dt = 1.25e-7
in_file = open('input.in', 'w')
for n in range(10, 1002, 5):
    nx.append(n)
    # Clear input file
    in_file.seek(0)
    in_file.write(f"{n}\n")
    in_file.write("0.0 1.0\n0.001 0.002")
    in_file.truncate()
    subprocess.call("./a.out > 1.txt", shell=True)
    # Read a number from out.txt
    out_file = open('1.txt', 'r')
    err_1.append(float(out_file.readline()))
    out_file.close()
in_file.close()
```

```python
nx = []
err_2 = []
# dt = 3.125e-6
in_file = open('input.in', 'w')
for n in range(10, 1002, 5):
    nx.append(n)
    # Clear input file
    in_file.seek(0)
    in_file.write(f"{n}\n")
    in_file.write("0.0 1.0\n0.001 0.002")
    in_file.truncate()
    subprocess.call("./a.out > 1.txt", shell=True)
    # Read a number from 1.txt
    out_file = open('1.txt', 'r')
    err_2.append(float(out_file.readline()))
    out_file.close()
in_file.close()
```

```python
# delete 2nd and 3rd element from each array
nx_del = np.delete(nx, [2, 3])
err_1_del = np.delete(err_1, [2, 3])
err_2_del = np.delete(err_2, [2, 3])
nx_log = np.log10(nx_del)
err_1_log = np.log10(err_1_del)
err_2_log = np.log10(err_2_del)
```

```python
plt.plot(nx_log, err_1_log, 'r', label='dt = 1.25e-7')
plt.plot(nx_log, err_2_log, 'b.', label='dt = 3.125e-6')
# plot a line y = -5.34x + 0.5
plt.plot(nx_log, -5.34*nx_log + 8, 'g', label='y = 8 - 4x')
# increase font size of axis labels and legend
plt.rcParams.update({'font.size': 12})
plt.xlabel('$log(nx)$')
plt.ylabel('$log(error)$')
plt.title("log plot of error vs nx")
plt.ylim(-10, 40)
plt.legend()
plt.show()
```

```python
i = 1
j = 0
print((err_2_log[i]-err_2_log[j])/(nx_log[i]-nx_log[j]))
print((err_1_log[i]-err_1_log[j])/(nx_log[i]-nx_log[j]))
```
