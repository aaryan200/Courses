```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# CFL = 0.3
arr = np.loadtxt('primitives_x_0159.dat')
```

```python
plt.plot(arr[:,0], arr[:,1]) # Density
plt.ylabel('Density')
plt.xlabel('x')
plt.title('Density vs. x')
plt.show()
```

```python
plt.plot(arr[:,0], arr[:,2]) # Velocity
plt.ylabel('Velocity')
plt.xlabel('x')
plt.title('Velocity vs. x')
plt.show()
```

```python
plt.plot(arr[:,0], arr[:,3]) # Pressure
plt.ylabel('Pressure')
plt.xlabel('x')
plt.title('Pressure vs. x')
plt.show()
```
