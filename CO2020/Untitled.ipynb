```python
import sympy as sp
```

```python
x = sp.Symbol("x")
y = sp.Symbol("y")
```

```python
T = sp.Function("T")(x,y)
T
```

```python
diff_eq = sp.Eq(T.diff(x,x)+T.diff(y,y), 0)
diff_eq
```

```python
rhb = sp.Function("r")(y)
rhb = 300*(1+0.2*(y/0.5 - 0.5))
```

```python
rhb
```

```python
ics = {T.subs(x,0):300, T.subs(x,1):rhb, T.diff(y).subs(y,0):0, T.diff(y).subs(y,0.5):0}
```

```python
ivp = sp.pdsolve(diff_eq, ics=ics)
```

```python
from sympy.abc import x, t
u, X, T = map(sp.Function, 'uXT')
eq = sp.Eq(sp.Derivative(u(x, t), x, 2), sp.Derivative(u(x, t), t, 2))
```

```python
eq
```

```python
a = sp.pde_separate(eq, u(x, t), [X(x), T(t)], strategy='mul')
```
