```python
a = np.loadtxt('1.txt')
plt.plot(a[:,0],a[:,1],'blue',label = 'Temperature of plate')
b = a[:,1].copy()
b.fill(a[-1,1])
plt.plot(a[:,0],b,'red', label = f'Furnace Temperature')
c = a[:,1].copy()
c.fill(a[-1,0])
plt.plot(c,a[:,1],'green', label = f'Equilibrium Time')
plt.xlabel('time')
plt.ylabel('Temperature (K)')
plt.xlim(0)
plt.title('Temperature vs Time for Initial Temperature = 300K')
plt.legend()
plt.show()
```
