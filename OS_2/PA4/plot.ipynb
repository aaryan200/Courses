```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
passg = [[10,11.5], [15,12.0], [20, 14.75], [25, 15.0], [30, 17.33], [35, 21.571], [40, 25.5], [45, 28.667], [50, 31]]
passg = np.array(passg)
plt.plot(passg[:,0], passg[:,1])
plt.xlabel('Number of Passengers')
plt.ylabel('Average Time')
plt.title('Average Time vs Number of Passengers')
plt.grid()
plt.show()
```

```python
cars = [[5, 110.0],[10, 58.0], [15, 34.0], [20, 28.5], [25, 20.0]]
cars = np.array(cars)
plt.plot(cars[:,0], cars[:,1])
plt.xlabel('Number of Cars')
plt.ylabel('Average Time')
plt.title('Average Time vs Number of Cars')
plt.grid()
plt.show()
```
