```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
threads = [10, 20, 30, 40, 50]

tas_worst = [8.8, 17.4, 27.2, 38.8, 45.8]
tas_avg = [0.88, 1.884, 2.93, 4.034, 5.002]

bcas_worst = [8, 17.2, 26.4, 36.6, 41.6] # cas_worst tha
cas_avg = [0.9, 1.883, 2.92, 4.0145, 5.0768]

cas_worst = [9, 19.2, 29.0, 39.4, 50.0] # bcas_worst tha
bcas_avg = [0.934, 2.025, 3.1366, 4.3595, 5.4192]
```

```python
plt.plot(threads, cas_worst, label='TAS')
plt.plot(threads, tas_worst, label='CAS')
plt.plot(threads, bcas_worst, label='BCAS')
plt.xlabel('Number of Threads')
plt.ylabel('Worst Waiting Time (s)')
# grid
plt.grid()
plt.title('Worst Waiting Time vs Number of Threads')
plt.legend()
plt.show()
```

```python
plt.plot(threads, tas_avg, label='TAS')
plt.plot(threads, cas_avg, label='CAS')
plt.plot(threads, bcas_avg, label='BCAS')
plt.xlabel('Number of Threads')
plt.ylabel('Average Waiting Time (s)')
# grid
plt.grid()
plt.title('Average Waiting Time vs Number of Threads')
plt.legend()
plt.show()
```
