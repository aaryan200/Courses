```python
# Openmp time vs thread for size = 25
thr = np.array([1,2,4,8,16,32,64])
tm = np.array([79,361,783,1536,2528, 5071,5500])
plt.plot(thr, tm,'b',label = 'Openmp')
for i in range(7):
    plt.plot(thr[i],tm[i],'ro-')

# Pthread time vs thread for size = 25
thr1 = np.array([1,2,4,8,16,32,64])
tm1 = np.array([500,600,700,1300,2000,3000,3400])
plt.plot(thr1, tm1,'k', label= 'Pthread')
plt.xlabel("Number of Threads")
plt.ylabel("Time taken")
for i in range(7):
    plt.plot(thr1[i],tm1[i],'go-')
plt.show()
```
