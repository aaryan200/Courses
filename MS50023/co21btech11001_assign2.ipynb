```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a0 = 0.529
Z = 1.0
```

```python
def psi_1s(r):
    """
    Wave function of 1s orbital
    """
    res = (1/np.sqrt(np.pi))*((Z/a0)**(1.5)) * np.exp(-Z*r/a0)
    return res
```

```python
def psi_2s(r):
    """
    Wave function of 2s orbital
    """
    res = (1/(4*np.sqrt(2*np.pi)))*((Z/a0)**(1.5)) * (2 - Z*r/a0) * np.exp(-Z*r/(2*a0))
    return res
```

```python
def psi_2p(r):
    """
    Wave function of 2p orbital
    """
    res = (1/(4*np.sqrt(2*np.pi)))*((Z/a0)**(1.5)) * (Z*r/a0) * np.exp(-Z*r/(2*a0))
    return res
```

```python
# Plot psi^2 for 1s, 2s, 2p orbitals
r = np.linspace(0, 4, 1000)
plt.plot(r, psi_1s(r)**2)
# Increase font size of xlabel and ylabel and title
plt.rcParams.update({'font.size': 16})
plt.title('1s orbital')
plt.xlabel('r (Bohr radius)')
plt.ylabel(r'$|\psi|^2$')
plt.ylim(0.0, 1.0)
plt.show()
```

```python
# Plot psi^2 for 1s, 2s, 2p orbitals
r = np.linspace(0, 8, 1000)
plt.plot(r, psi_2s(r)**2)
# Increase font size of xlabel and ylabel and title
plt.rcParams.update({'font.size': 16})
plt.title('2s orbital')
plt.xlabel('r (Bohr radius)')
plt.ylabel(r'$|\psi|^2$')
plt.ylim(0.0, 0.02)
plt.show()
```

```python
# Plot psi^2 for 1s, 2s, 2p orbitals
r = np.linspace(0, 8, 1000)
plt.plot(r, psi_2p(r)**2)
# Increase font size of xlabel and ylabel and title
plt.rcParams.update({'font.size': 16})
plt.title('2p orbital')
plt.xlabel('r (Bohr radius)')
plt.ylabel(r'$|\psi|^2$')
plt.ylim(0.0, 0.1)
plt.show()
```

```python
r = np.linspace(0, 4, 1000)
# Plot 4*pi*r^2*psi^2 for 1s, 2s, 2p orbitals
plt.plot(r, 4*np.pi*r**2*psi_1s(r)**2)
# Increase font size of xlabel and ylabel and title
plt.rcParams.update({'font.size': 16})
plt.title('1s orbital')
plt.xlabel('r (Bohr radius)')
plt.ylabel(r'$4\pi r^2|\psi|^2$')
plt.ylim(0.0, 1.5)
plt.show()
```
