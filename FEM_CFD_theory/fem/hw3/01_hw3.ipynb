```python
import numpy as np
import matplotlib.pyplot as plt

def gauss_integration(f, deg, a, b):
    """
    Gauss integration of function f(x) from a to b with degree deg
    """
    # n_gp >= (deg + 1) / 2
    n_gp = (deg + 2) // 2
    zeta, W = np.polynomial.legendre.leggauss(n_gp)
    x = (a + b)/2 + ((b-a)/2)*zeta
    fx = f(x)
    J = (b-a)/2
    return float(J*np.dot(W, fx))
```

```python
gauss_integration(lambda x: (np.cos(np.pi * x)**2), 11, -1, 1)
```

```python
vals = []
for deg in range(2, 15):
    vals.append(gauss_integration(lambda x: (np.cos(np.pi * x)**2), deg, -1, 1))

plt.plot(range(2, 15), vals)
plt.grid()
plt.axhline(y=1, color='r', linestyle='--')
plt.show()
```

```python
vals = []
for deg in range(2, 5):
    vals.append(gauss_integration(lambda x: (x/(x**2 + 1)), deg, -1, 1))

plt.plot(range(2, 5), vals)
plt.grid()
plt.axhline(y=0, color='r', linestyle='--')
plt.show()
```

```python
import scipy as sp
def std_normal_integ(x):
    """
    Returns the integral from -inf to x of the standard normal distribution
    """
    return sp.stats.norm.cdf(x)
```

```python
std_normal_integ(np.sqrt(2)) - std_normal_integ(1/np.sqrt(2))
```

```python
1.0 - std_normal_integ(1/np.sqrt(2))
```
