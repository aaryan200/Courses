```python
# Create a Fibonacci series
def fib(n):
    a, b = 0, 1
    for i in range(n):
        a, b = b, a+b
    return a
```

$k(x) = -x$ if $x<0$

$k(x) = x^3$ if $0\leq x < 1$

$k(x) = x^2$ if $1\leq x < 2$

$k(x) = 4$ otherwise

```python
x=np.linspace(-1,3,9)
choices=[ x>=2, x>=1, x>=0, x<0 ]
outcomes=[ 4.0, x**2, x**3, -x ]
k=np.select(choices, outcomes)
print(x,k)
```

The first rule of broadcasting is that, if the arrays do not have the same number of
dimensions, then a $1$ will be repeatedly prepended to the shapes of the smaller
arrays until all the arrays have the same number of axes.

The second rule of broadcasting ensures that arrays with a size of 1 along a particular
dimension or axis act as if they had the size of the array with the largest size along
that dimension. The value of the array element is assumed to be the same along
that dimension for the “broadcasted” array

```python
# * is element wise multiplication
# @ is matrix multiplication
x=np.array([[0,1,2,3],[10,11,12,13],[20,21,22,23]])
r=np.array([[2,3,4,5]])
c=np.array([[5],[6],[7]])
print("x=\n", x)
print("r = \n",r)
print("2*x = \n", 2*x)
print("r*x = \n", r*x)
print("x*r = \n", x*r)
print("c*x = \n", c*x)
print("x@r.T = \n", x@(r.T))
print("x.(r.T) = \n", np.dot(x,r.T))
x.shape, r.shape, c.shape
```

```python
import numpy as np
import matplotlib.pyplot as plt

maxIter = 2000

Lx = Ly = 20
dx = 1

# Boundary conditions
Ttop = 100
Tbottom = 30
Tleft = 30
Tright = 30

Tguess = 50
colorinterpolation = 50
colourMap = plt.cm.jet

# Set coordinates
X, Y = np.meshgrid(np.arange(0,Lx,1), np.arange(0,Ly,1))
T = np.empty((Lx, Ly))
T.fill(Tguess)

# Set Boundary conditions
T[(Ly-1):, :] = Ttop
T[:1, :] = Tbottom
T[:, (Lx-1):] = Tright
T[:, :1] = Tleft

for it in range(0, maxIter):
    for i in range(1, Lx-1, dx):
        for j in range(1, Ly-1, dx):
            T[i, j] = 0.25 * (T[i+1][j] + T[i-1][j] + T[i][j+1] + T[i][j-1])

plt.title("Temperature field")
plt.contourf(X, Y, T, colorinterpolation, cmap=colourMap)
plt.colorbar()
plt.show()
```

```python
import numpy as np
x1 = 0; alpha = 30
x2 = 5; beta = 100
npts = 1000
b = np.zeros((npts, 1));
b[0] = -alpha
b[-1] = -beta
A = np.zeros((npts, npts));
for i in range(npts):
    for j in range(npts):
        if j == i: 
            A[i,j] = -2
        elif j == i - 1:
            A[i,j] = 1
        elif j == i + 1:
            A[i,j] = 1

Y = np.linalg.solve(A,b)

x = np.linspace(x1, x2, npts + 2)
y = np.hstack([alpha, Y[:,0], beta])

import matplotlib.pyplot as plt

plt.plot(x, y)

plt.plot(x, alpha + (beta - alpha)/(x2 - x1) * x, 'r--')

plt.xlabel('X')
plt.ylabel('T(X)')
plt.legend(('finite difference', 'analytical soln'), loc='best')
```
